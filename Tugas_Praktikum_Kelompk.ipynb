```python
import pandas as pd

dpath = '/content/drive/MyDrive/ML/wbc.csv'
df = pd.read_csv(dpath)
df.head()
```

```python
df.head(10)
```

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df = df[['id', 'diagnosis', 'radius_mean', 'compactness_mean', 'perimeter_worst']]
df.head()
```

2. Lakukan proses encoding pada kolom "diagnosis"

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding
```

```python
df.head()
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
df['diagnosis'] = std.fit_transform(df[['diagnosis']])
df['radius_mean'] = std.fit_transform(df[['radius_mean']])
df['compactness_mean'] = std.fit_transform(df[['compactness_mean']])
df['perimeter_worst'] = std.fit_transform(df[['perimeter_worst']])
```

```python
df.head()
```

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# Split data
from sklearn.model_selection import train_test_split

# Pilih kolom yang akan digunakan
selected_columns = ['id', 'diagnosis', 'radius_mean', 'compactness_mean', 'perimeter_worst']
df_selected = df[selected_columns]

# Lakukan stratified split data latih dan data uji dengan rasio 80:20
df_train, df_test = train_test_split(df_selected, test_size=0.2, random_state=0, stratify=df_selected['diagnosis'])

# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{df_selected["diagnosis"].value_counts()}')
print('\n')
print(f'Jumlah label data train:\n{df_train["diagnosis"].value_counts()}')
print('\n')
print(f'Jumlah label data test:\n{df_test["diagnosis"].value_counts()}')
```
